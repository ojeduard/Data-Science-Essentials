# Our World Connected Project

```python
import pandas as pd
import matplotlib.pyplot as plt

internet =  pd.read_csv('world-connected-project/world-internet-users.csv')
internet
```

```python
# En que anho exceden los 100 millones de usuarios
exceeds_100M = internet.query('internet_users > 100e6')
exceeds_100M.head(1)
```

```python
# World population
population = pd.read_csv('world-connected-project/historical-world-population.csv')
population
```

```python
# Merge both tables
merged_data = internet.merge(population, on='year', how='left')
merged_data
```

```python
# To avoid the NaN values we can use dropna()
merged_data = merged_data.dropna()
merged_data
```

```python
# This way is possible calculate the percentage of internet users
merged_data['percent'] = merged_data.eval('internet_users / population * 100')
merged_data['percent'] = merged_data['percent'].round(2)
merged_data
```

```python
# let's plot the percentage of internet users
plt.plot(merged_data['year'], merged_data['percent'])
plt.axhline(50, color='red', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Percent Connected')

```

```python
over_half_connected = merged_data.query('percent >= 50')
over_half_connected.head(1)
```

```python
import pandas as pd
features = pd.read_csv('mondrian-art-project/mondrian-painting-features.csv')
features
```

```python
# There paintings id are not unique. Let's which have the same as the first one
features.query('painting_id == "b104"')
```

```python
# There info we are missing, so let[s load the info from other file
painting_info = pd.read_csv('mondrian-art-project/mondrian-painting-info.csv')
painting_info
```

```python
#For this project, we've provided the following function to draw Mondrian paintings 
# from data
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
def draw_mondrian( painting_id ):
    rects = features.query('painting_id == @painting_id')
    total_width = rects.eval("x + width").max()
    total_height = rects.eval("y + height").max()
    
    fig, ax = plt.subplots()
    
    for (idx, row) in rects.iterrows():
        x, y, w, h, rgb = row[['x','y','width','height','rgb']]
        patch = mpatches.Rectangle((x, y), w, h, facecolor=rgb)
        ax.add_patch(patch)
    
    ax.axis([0, total_width, 0, total_height])
    ax.set_aspect('equal')
    ax.axis('off')
    fig.text(0.5, 0.01, painting_id, ha="center", fontsize=14)
    fig.show()
    
```

```python
draw_mondrian('b104')
```

```python
# Let's see with other painting ids
draw_mondrian('b106')
draw_mondrian('b107')
```

```python
# Gruouping the features by painting id to see how many features each painting has
sizes = features.groupby('painting_id').size()
sizes
```

```python
# Turn series to DataFrame
complexity_df = sizes.reset_index(name=('complexity'))
complexity_df
```

```python
# Let's merge
# We want to plot complexity over time, 
# but we have year and complexity in separate dataframes

painting_info = painting_info.merge(complexity_df, on='painting_id', how='left')
painting_info
```

```python
# Let's plot the complexity over time
import matplotlib.pyplot as plt
plt.scatter(painting_info['year'], painting_info['complexity'])
plt.xlabel('Year')
plt.ylabel('Complexity')
```

```python
class Team:
    def __init__(self, name, members):
        self.__name = name
        self.members = members
    
    def add_member(self, mem):
        self.members.append(mem)
    
    def get_name(self):
        return self.__name
    
    def show_team(self):
        print(f"This is a soccer team called {self.get_name} and it members are:")
        for member in self.members:
            print(member)

class BasketballTeam(Team):
    def __init__(self, name, members):
        super().__init__(name, members)


    def show_team(self):
        print(f"This is a basketball team called {self.get_name} and it members are:")
        for member in self.members:
            print(member)


basketball_team = BasketballTeam("Los campeones", ["Alice", "Bob", "Charlie"])
print(basketball_team.get_name())  # Output: Los campeones
basketball_team.show_team()

```
