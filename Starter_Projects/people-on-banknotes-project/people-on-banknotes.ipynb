# People on Banknotes

Whose faces appear on banknotes?

The file `people-on-banknotes.csv` contains data about individuals featured on banknotes from 38 countries. This dataset spans all 22 subregions and sub-subregions of the world, as defined by the United Nations Statistics Division's geoscheme.

It profiles 241 people, detailing their occupations and the year they first appeared on a banknote. Additionally, it includes their year of death — or `NaN` if they were still alive when the dataset was compiled.

Most banknotes were issued after the featured individual’s death. The column `first_death_diff` calculates the difference between the year of their first appearance on a banknote and their year of death (or remains `NaN` if the person was still living at the time of curation).

```python
import pandas as pd

df = pd.read_csv('people-on-banknotes.csv')
df
```

### Quick cleaning

The same person can appear on multiple banknotes. Below we drop the `value` column and remove duplicate people.

```python
df = df.drop(columns=['value'])
df = df.drop_duplicates(subset="name")
df
```

### Project Ideas

- What proportion of individuals featured are male versus female?
	- Hint: Use `value_counts(normalize=True)` to calculate percentages.

- Are writers or politicians more commonly depicted?

- What percentage of featured individuals are musicians?

- What percentage of banknotes were issued before the person’s death?
	- Hint: Look for negative values or NaN in `first_death_diff`.

- Who is the oldest historical figure in the dataset?

- Which countries feature the oldest historical figures on their banknotes?
	- Hint: Group by country and aggregate the year of death using the median. Sort the results.

- What percentage of individuals died at least 100 years before appearing on a banknote?

- Which individuals appeared on a banknote just one year after their death?


```python
# YOUR CODE HERE (add additional cells as needed)
df['death'] = pd.to_numeric(df['death'], errors='coerce')
# The number of people on banknotes per gender
gender_proportion = round(df['gender'].value_counts(normalize=True) * 100, 2)
print(gender_proportion.to_string())


"""
Esta es una manera de hacerlo pero dejaremos la mas pythonica para el final.


writer_count = len(df[df['occupation'] == 'Writer'])
politician_count = df[df['occupation'] == 'Politician'].shape[0]

most_common_occupation = 'Writer' if writer_count > politician_count else 'Politician'

"""

# The most common occupation on banknotes im grouped by occupation
occupations = ['Writer', 'Politician']
counts = {occ: df[df['occupation'] == occ].shape[0] for occ in occupations} # dictionary comprehension
most_common_occupation = max(counts, key=counts.get)

# The percentage of musicians on banknotes
musician_count_percentage = round(df[df['occupation'] == 'Musician'].shape[0] / df.shape[0] * 100, 2)

# The percentage of banknotes issued before the person's death
banknotes_before_death = round(df.query('first_death_diff < 0 or first_death_diff.isna()').shape[0] / df.shape[0] * 100, 2)

# The oldest person on banknotes
oldest_figure = df[df['death'].notna()].loc[df['death'].idxmin(), 'name']

# The oldest countries by death count
oldest_countries = df.groupby('country').agg({'death' : 'median'}).reset_index().sort_values(by='death', ascending=True).head(5)

# The percentage of people on banknotes who died at least 100 years before appearing on the banknotes
diff_100_years = round(df[df['first_death_diff'] >= 100].shape[0] / df.shape[0] * 100, 2)

# Individuals who appeared on a banknote just one year after their death
one_year_after_death = df.query('first_death_diff == 1')


print(f'\nThe most common occupation on banknotes is: {most_common_occupation}')
print(f'The percentage of musicians on banknotes is: {musician_count_percentage}%')
print(f'The percentage of banknotes issued before the person\'s death is: {banknotes_before_death}%')
print(f'The oldest person on banknotes is: {oldest_figure}')
print(f'The oldest countries by death count are:\n{oldest_countries}')
print(f'The percentage of people on banknotes who died at least 100 years before appearing on the banknotes is: {diff_100_years}%')
print (f'Individuals who appeared on a banknote just one year after their death:\n{one_year_after_death[["name", "country", "death"]].sort_values(by="death", ascending=False)}')

```
