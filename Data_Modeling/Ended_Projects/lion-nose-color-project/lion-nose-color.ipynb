# Lion Nose Color

Can you estimate a lion's age from a photograph of its nose?

Lions are born with pink noses that gradually darken with age, developing black freckles until they eventually turn completely black.

The dataset `lion-nose-color.csv` contains data on 91 lions from the Serengeti, each monitored by researchers since birth. Because their exact ages are known, researchers can analyze photographs to determine the percentage of black on their noses and assess how well nose coloration predicts age.

```python
import pandas as pd
import matplotlib.pyplot as plt
from linear_model import LinearModel

df = pd.read_csv('lion-nose-color.csv')
df.head()
```

# Project Ideas

```python
# YOUR CODE HERE (add additional cells as needed)

# Linear model API

# linear = LinearModel()
# linear.fit(x, y)
# linear.predict(x)
# linear.plot_model(x_min, x_max, color="black")
# linear.print_model_info(self):

# Firstable we create a scatter plot to visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(df['proportion_black'], df['age'])
plt.xlabel('Color Score')
plt.ylabel('Age')
plt.title('Lion Age vs Color Score')
plt.show()
```

```python
# Now we create and fit the linear model
linear = LinearModel('Age vs Color Score')
linear.fit(df['proportion_black'], df['age'])
linear.print_model_info()
linear.plot_model(0, 100, color="red")
plt.scatter(df['proportion_black'], df['age'])
plt.xlabel('Color Score')
plt.ylabel('Age')
plt.title('Lion Age vs Color Score with Linear Model')
plt.show()
```

```python
lion_age_nose_50 = linear.predict(50)
lion_age_nose_100 = linear.predict(100)
print(f'Predicted age for lion with color score 50: {lion_age_nose_50.round(2)}')
print(f'Predicted age for lion with color score 100: {lion_age_nose_100.round(2)}')

"""
RECORDAR QUE LA VARIABLE X LA INDEPENDIENTE Y LA Y ES LA DEPENDIENTE, ES DECIR, LA EDAD VA A DEPENDER DEL COLOR DE LA NARIZ.
"""

```
